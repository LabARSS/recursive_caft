```python
from pathlib import Path

import pandas as pd

from core.utils.seed import set_seed

set_seed()

TEST_ALLOCATION = 0.2
SPLIT_OUT_PATH = Path("../../data/out/splits/single_token_entropy/qwen_3b/")

ans_col = "entropy_ans_qwen2"
entropy_col = "entropy_value_qwen2"

df = pd.read_csv(
    "../../data/out/single_token_entropy/mmlu_qwen_3b.tsv",
    sep="\t",
    header=0,
    dtype={ans_col: "str"},
)
```

```python
from core.utils.splitter import split_into_even_chunks

chunks = split_into_even_chunks(df, entropy_col, ans_col, chunk_cnt=6)
```

```python
print("Chunks: ", len(chunks))

for chunk in chunks:
    print("Chunk len: ", len(chunk))
    print("Chunk sample")
    print(chunk.head()[entropy_col])
```

```python
from core.utils.splitter import split_chunk_into_train_test

for chunk_i, chunk in enumerate(chunks):
    print("Chunk : ", chunk_i)

    train_df, test_df = split_chunk_into_train_test(chunk, TEST_ALLOCATION)
    print("Train len: ", len(train_df))
    print("Test len: ", len(test_df))

    test_df.to_csv(str(SPLIT_OUT_PATH.joinpath(f"group{chunk_i}_test.tsv").resolve()), sep="\t", index=False)
    train_df.to_csv(str(SPLIT_OUT_PATH.joinpath(f"group{chunk_i}_train.tsv").resolve()), sep="\t", index=False)
```
